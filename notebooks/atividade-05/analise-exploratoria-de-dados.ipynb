## Insights:

```python
ax = sns.histplot(data=df, x='Rings')

#UTILIZE LINHAS VERTICAIS E HORIZONTAIS
sns.set_style("whitegrid")

#MELHORE O TITULO E OS LABELS DOS EIXOS
ax = sns.histplot(data=df, x='Rings')
plt.ylabel('Quantidade')
plt.xlabel('Anéis')
plt.title('Distribuição por quantidade de aneis')
# Adicione uma linha a origem e remova as bordas:
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```

```python
ax = sns.histplot(data=df, x='Viscera weight')

#UTILIZE LINHAS VERTICAIS E HORIZONTAIS
sns.set_style("whitegrid")

#MELHORE O TITULO E OS LABELS DOS EIXOS
ax = sns.histplot(data=df, x='Viscera weight')
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa')
# Adicione uma linha a origem e remova as bordas:
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```

```python
ax = sns.histplot(data=df, x='Shell weight')

#UTILIZE LINHAS VERTICAIS E HORIZONTAIS
sns.set_style("whitegrid")

#MELHORE O TITULO E OS LABELS DOS EIXOS
ax = sns.histplot(data=df, x='Shell weight')
plt.ylabel('Quantidade')
plt.xlabel('tamanho')
plt.title('Distribuição por tamanho da cascca')
# Adicione uma linha a origem e remova as bordas:
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```
